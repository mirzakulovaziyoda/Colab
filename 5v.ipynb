1️⃣ Bosqich: Loyiha nomi va maqsadi

Nom: O‘zbekistonning xalqaro savdo oqimlari tahlili
Maqsad: Eksport va import hajmlarini, savdo balansini va asosiy hamkor mamlakatlar bo‘yicha taqsimotini o‘rganish hamda kelgusi prognozlarni ishlab chiqish.

2️⃣ Bosqich: Ma’lumotlarni olish

Manbalar: World Bank WITS, UN Comtrade, IMF DOTS, O‘zbekiston Davlat statistika qo‘mitasi.

Dastlab sinov uchun kichik demo dataset ishlatamiz.

```python
import pandas as pd

data = {
    "Year": [2018, 2019, 2020, 2021, 2022, 2023],
    "Export": [12.4, 13.1, 12.0, 15.3, 17.8, 19.2],  # mlrd $
    "Import": [21.0, 23.5, 22.1, 25.4, 28.7, 30.1],  # mlrd $
}
df = pd.DataFrame(data)
df
```

3️⃣ Bosqich: Ma’lumotlarni tozalash va tayyorlash

Yillarni int turiga o‘tkazamiz

Savdo balansini qo‘shamiz

```python
df["Year"] = df["Year"].astype(int)
df["Trade_Balance"] = df["Export"] - df["Import"]
df
```

1️⃣2️⃣ Bosqich: Linear Regression asosida prognoz

```python
from sklearn.linear_model import LinearRegression
import numpy as np

X = df[["Year"]]
y = df["Export"]

model = LinearRegression()
model.fit(X, y)

future_years = np.array(range(2024, 2030)).reshape(-1,1)
forecast = model.predict(future_years)

plt.figure(figsize=(8,5))
plt.plot(df["Year"], df["Export"], marker='o', label="Haqiqiy Eksport")
plt.plot(future_years, forecast, marker='x', linestyle="--", label="Prognoz")
plt.title("Eksport hajmi prognozi (Linear Regression)")
plt.xlabel("Yil")
plt.ylabel("Eksport (mlrd $)")
plt.legend()
plt.show()
```

🌍 World Bank API’dan ma’lumot olish (O‘zbekiston savdo oqimlari)

```python
import pandas as pd
import requests

# World Bank API indikatorlari:
# NE.EXP.GNFS.CD → Eksport (Goods & Services, current US$)
# NE.IMP.GNFS.CD → Import (Goods & Services, current US$)

country = "UZB"  # O‘zbekiston
indicators = {
    "Export": "NE.EXP.GNFS.CD",
    "Import": "NE.IMP.GNFS.CD"
}

data = {}

for name, code in indicators.items():
    url = f"http://api.worldbank.org/v2/country/{country}/indicator/{code}?format=json&per_page=100"
    r = requests.get(url).json()[1]
    data[name] = {item["date"]: item["value"] for item in r if item["value"] is not None}

# DataFrame shaklida
df = pd.DataFrame(data).sort_index()
df.index = df.index.astype(int)  # yilni int ga aylantiramiz
df = df.loc[2010:2024]  # kerakli davr

# Savdo balansi
df["Trade_Balance"] = df["Export"] - df["Import"]

df.head(10)
```

📊 Vizualizatsiya

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.plot(df.index, df["Export"]/1e9, marker='o', label="Eksport (mlrd $)")
plt.plot(df.index, df["Import"]/1e9, marker='o', label="Import (mlrd $)")
plt.plot(df.index, df["Trade_Balance"]/1e9, marker='o', label="Savdo balansi (mlrd $)")
plt.title("O‘zbekiston xalqaro savdo oqimlari (2010–2024)")
plt.xlabel("Yil")
plt.ylabel("mlrd AQSh dollari")
plt.legend()
plt.grid(True)
plt.show()
```

✍️ Umumiy xulosa – O‘zbekistonning xalqaro savdo oqimlari (2010–2024) bo‘yicha

Eksport va import hajmi:

Prognoz:
